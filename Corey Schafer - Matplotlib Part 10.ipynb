```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('fast')
```

```python
import numpy as np

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
```

```python
plt.hist(ages, bins=5)
plt.xlabel('Ages')
plt.title('# by Age Groups')
plt.show()
```

```python
bins = [10,20,30,40,50,60]

plt.hist(ages, bins=bins)
plt.xlabel('Ages')
plt.title('# by Age Groups')
plt.show()
```

```python
df = pd.read_csv('/home/noor/Documents/code_snippets-master/Python/Matplotlib/06-Histograms/data.csv')
```

```python
df.head(3)
```

```python
bins = [i for i in range(10,110,10)]

plt.hist(df.Age, bins=bins)
plt.xlabel('Ages')
plt.ylabel('Number of Respondents')
plt.title('# by Age Groups')
plt.show()
```

```python
bins = [i for i in range(10,110,10)]

plt.hist(df.Age, bins = bins, edgecolor = 'black', log = True)
plt.xlabel('Ages')
plt.ylabel('Number of Respondents')
plt.title('# by Age Groups')
plt.show()
```

```python
bins = [i for i in range(10,110,10)]

plt.hist(df.Age, bins = bins, edgecolor = 'black', log = True)
plt.xlabel('Ages')
plt.ylabel('Number of Respondents')
plt.axvline(df.Age.median(), color='green')
plt.title('# by Age Groups')
plt.show()
```
