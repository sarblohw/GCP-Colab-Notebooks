```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('/home/noor/Documents/code_snippets-master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv')
```

```python
plt.style.use('seaborn-pastel')
```

```python
import numpy as np
x = [i for i in np.random.randint(10,75,25)]
y = [i for i in np.random.randint(10,75,25)]
```

```python
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter')
plt.show()
```

```python
plt.scatter(x, y, s=100, c='green', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter')
plt.grid()
plt.show()
```

```python
plt.scatter(x, y, s=100, marker='o', edgecolor='grey', linewidth=1, alpha=0.75)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
colors = [i for i in np.random.randint(1,10,25)]

plt.scatter(x, y, s=100, c=colors, cmap='GnBu', marker='x')
plt.colorbar(label='Satisfaction Score')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter')
plt.grid()
plt.show()
```

```python
colors = [i for i in np.random.randint(1,10,25)]
sizes = [i for i in np.random.randint(20,500,25)]

plt.scatter(x, y, s=100, c=colors, cmap='GnBu', marker='x', sizes=sizes)
plt.colorbar(label='Satisfaction Score')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter')
plt.grid()
plt.show()
```

```python
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75, cmap='GnBU')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.title('Likes by Views')
plt.show()
```

```python
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75, c=ratio, cmap='summer')
plt.colorbar(label='Like/Dislike Ratio')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.title('Likes by Views')
plt.show()
```
