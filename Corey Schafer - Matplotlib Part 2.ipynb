```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
age = [i for i in range(25, 35)]
devs = [int(str(i).replace(str(i)[-2:],'00')) for i in np.random.randint(24000, 78000, 10)]
jsdevs = [int(str(i).replace(str(i)[-2:],'00')) for i in np.random.randint(24000, 78000, 10)]
pydevs = [int(str(i).replace(str(i)[-2:],'00')) for i in np.random.randint(24000, 78000, 10)]
godevs = [int(str(i).replace(str(i)[-2:],'00')) for i in np.random.randint(24000, 78000, 10)]
```

```python
%matplotlib inline
```

```python
plt.style.use('ggplot')

plt.bar(age, devs, label='All Devs')
plt.plot(age, jsdevs, label='JavasScript')
plt.plot(age, pydevs, 'x--c', label='Python', linewidth=3)
plt.plot(age, godevs, label='Go')
plt.xlabel('Age')
plt.ylabel('Compensation')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.title('Age vs Compensation for Developers')
plt.show()
```

```python
#trying to plot a clustered bar chart with no indexes or width defined will just overlap the plots

plt.bar(age, devs, label='All_Devs', color='grey')
plt.bar(age, jsdevs, label='JS')
plt.bar(age, pydevs, label='Py')
plt.bar(age, godevs, label='Go')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.title('Comp vs Age')
plt.show()
```

```python
indexes = np.arange(len(age))
width = 0.25

plt.bar(indexes, devs, label='All_Devs', color='grey', width=width)
plt.bar(indexes+width, jsdevs, label='JS', width=width)
plt.bar(indexes-width, pydevs, label='Py', width=width)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.title('Comp vs Age')
plt.show()
```
