```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
age = [i for i in range(25, 35)]
devs = [int(str(i).replace(str(i)[-2:],'00')) for i in np.random.randint(24000, 78000, 10)]
jsdevs = [int(str(i).replace(str(i)[-2:],'00')) for i in np.random.randint(24000, 78000, 10)]
pydevs = [int(str(i).replace(str(i)[-2:],'00')) for i in np.random.randint(24000, 78000, 10)]
```

```python
plt.style.use('seaborn-notebook')
plt.bar(age, devs, label='All Developers')
plt.plot(age, jsdevs, 'x--c', label='Javascript Developers', linewidth=2)
plt.plot(age, pydevs, 'o-.y', label='Python Developers', linewidth=3)
plt.xlabel('Age')
plt.ylabel('Compensation')
plt.xticks(np.arange(25,35,3))
plt.yticks([25000, 50000, 75000], labels=['25K', '50K', '75K'])
plt.legend()
plt.tight_layout()
plt.title('Age vs Compensation for Developers')
plt.show()
```

```python
devs_srt = sorted(devs)

plt.barh(age, devs_srt, label='All Developers')
plt.xlabel('Compensation')
plt.ylabel('Age')
plt.legend()
plt.tight_layout()
plt.title('Age vs Compensation for Developers')
plt.show()
```

```python
plt.bar(age, devs, label='All Developers', width = 0.25, color='pink')
plt.bar([x+0.25 for x in age], jsdevs, label='Javascript Developers', width = 0.25, color='cyan')
plt.bar([x-0.25 for x in age], pydevs, label='Python Developers', width = 0.25, color='orange')
plt.xlabel('Age')
plt.ylabel('Compensation')
plt.legend()
plt.tight_layout()
plt.title('Age vs Compensation for Developers')
plt.show()
```
