```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv('survey_results_public.csv', usecols=['LanguageWorkedWith'])
```

```python
import csv
c = Counter()
with open('survey_results_public.csv') as file:
    for row in csv.DictReader(file):
        c.update(row['LanguageWorkedWith'].split(';'))
c.most_common(15)
```

```python
{k:v for k,v in sorted(c.items(), key = lambda item: item[1], reverse=True)}
```

```python
c = Counter()

with open('survey_results_public.csv') as f:
    
    reader = csv.DictReader(f)
    
    for row in reader:
        c.update(row['LanguageWorkedWith'].split(';'))

    languages = []
    popularity = []

    for i in c.most_common(15):
        languages.append(i[0])
        popularity.append(i[1])

plt.figure(figsize=(10,6))
plt.style.use('fivethirtyeight')
plt.barh(languages, popularity)
plt.xlabel('Language')
plt.ylabel('Popularity')
plt.annotate('Sumi', xy = (55000, 'Ruby'))
plt.title('Popularity of Languages')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
c = Counter()

with open('survey_results_public.csv') as f:
    
    reader = csv.DictReader(f)
    
    for row in reader:
        c.update(row['LanguageWorkedWith'].split(';'))
    
    c = sorted(c.items(), key=lambda x: x[1], reverse=False)

    languages = []
    popularity = []

    for i in c[:16]:
        languages.append(i[0])
        popularity.append(i[1])

plt.figure(figsize=(10,6))
plt.style.use('fivethirtyeight')
plt.barh(languages, popularity)
plt.xlabel('Language')
plt.ylabel('Popularity')
plt.annotate('Sumi', xy = (55000, 'Ruby'))
plt.title('Popularity of Languages')
plt.grid(False)
plt.tight_layout()
plt.show()
```
