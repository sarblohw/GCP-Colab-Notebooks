```python
slices = [60, 40]
labels = ['sixty', 'forty']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor' : 'grey'})
plt.title('Pie')
plt.tight_layout()
plt.show()
```

```python
slices = [60, 40, 30, 20]
labels = ['Sixty', 'Forty', 'Thirty', 'Twenty']

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

plt.pie(slices, labels=labels, wedgeprops={'edgecolor' : 'grey'}, startangle=90, shadow=True, explode=[0,0,0.2,0])
plt.title('Pie Chart', fontdict=font)
plt.tight_layout()
plt.show()
```

```python
import csv
from collections import Counter

c = Counter()

with open('survey_results_public.csv') as f:
    for i in csv.DictReader(f):
        c.update(i['LanguageWorkedWith'].split(';'))

c
```

```python
top5lang = [i[0] for i in c.most_common(5)]
popularity = [i[1] for i in c.most_common(5)]
```

```python
popularity
```

```python
top5lang
```

```python
plt.pie(popularity, labels = top5lang, startangle = 0, shadow = True, wedgeprops = {'edgecolor': 'grey'}, explode = [0,0,0,0.2,0], autopct = '%.1f%%')
plt.title('5 most popular languages', fontdict=font)
plt.tight_layout()
plt.show()
```
