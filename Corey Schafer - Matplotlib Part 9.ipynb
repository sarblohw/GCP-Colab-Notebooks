```python
#esports gameplay data example
import numpy as np
minute = [i for i in np.arange(1,10)]
```

```python
minute
```

```python
player1 = [1,1,2,3,4,5,5,6,6]
player2 = [1,1,1,3,4,5,5,5,5]
player3 = [1,2,3,3,3,4,5,8,8]
```

```python
#stack chart
```

```python
labels = ['P1', 'P2', 'P3']

plt.stackplot(minute, player1, player2, player3, labels=labels)
plt.title('Minute on minute play')
plt.tight_layout()
plt.legend(loc= (0.04, 0.75))
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('/home/noor/Documents/code_snippets-master/Python/Matplotlib/05-Fill_Betweens/data.csv')

df.columns
```

```python
ages = df['Age']
devs_all = df['All_Devs']
py = df['Python']
js = df['JavaScript']
```

```python
df.head(3)
```

```python
plt.style.use('seaborn-pastel')
plt.figure(figsize=(12,7))
plt.plot(ages, devs_all, '--', label='All Devs')
plt.plot(ages, py, '-', label='Python')
plt.xlabel('Age')
plt.ylabel('Median Salary USD')
plt.legend()
plt.title('Age by Median Salaries')
plt.show()
```

```python
plt.style.use('seaborn-pastel')
plt.figure(figsize=(12,7))
plt.plot(ages, devs_all, '--', label='All Devs')
plt.plot(ages, py, '-', label='Python')
plt.fill_between(ages, py, alpha=0.20)
plt.xlabel('Age')
plt.ylabel('Median Salary USD')
plt.legend()
plt.title('Age by Median Salaries')
plt.show()
```

```python
df.loc[:, 'All_Devs'].median()
```

```python
plt.style.use('seaborn-pastel')
plt.figure(figsize=(12,7))
plt.plot(ages, devs_all, '--', label='All Devs')
plt.plot(ages, py, '-', label='Python')
plt.fill_between(ages, py, df['All_Devs'].median(), alpha=0.20)
plt.xlabel('Age')
plt.ylabel('Median Salary USD')
plt.legend()
plt.title('Age by Median Salaries')
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'grey',
        'weight': 'normal',
        'size': 16,
        }

plt.figure(figsize=(12,9))
plt.plot(ages, devs_all, '--', label = 'All Devs')
plt.plot(ages, py, label = 'Python')
plt.fill_between(ages, py, df.All_Devs.median(), where = py>df.All_Devs.median(), alpha = 0.20, interpolate=True)
plt.fill_between(ages, py, df.All_Devs.median(), where = py<df.All_Devs.median(), alpha = 0.20, color = 'grey', interpolate=True)
plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median Salary by Age', fontdict=font)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,9))
plt.plot(ages, devs_all, '--', label = 'All Devs')
plt.plot(ages, py, label = 'Python')
plt.fill_between(ages, py, devs_all, where = py>devs_all, alpha = 0.20, interpolate=True, label='Above Avg')
plt.fill_between(ages, py, devs_all, where = py<devs_all, alpha = 0.20, color = 'grey', interpolate=True, label='Below Avg')
plt.xlabel('Ages')
plt.ylabel('Median Salary USD')
plt.title('Median Salary by Age', fontdict=font)
plt.tight_layout()
plt.legend()
plt.show()
```
