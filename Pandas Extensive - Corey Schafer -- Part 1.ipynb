#### Part 4 - Filtering

```python
multiple_people = {
    'fname' : ['Suminder', 'Deepinder', 'Nimrat'],
    'lname' : ['Singh', 'Kaur', 'Kaur'],
    'email' : ['suminder.singh@ex.com', 'dkaur@ex.com', 'nkaur@ex.com']
}

df_tst = pd.DataFrame(multiple_people)
```

```python
df_tst['lname'] == 'Singh' # filter mask
```

```python
filt = (df_tst['lname'] == 'Singh')
df_tst[filt]
```

```python
df_tst[df_tst['lname'] == 'Singh']
```

```python
df_tst.loc[filt]
```

```python
df_tst.loc[filt, 'email']
```

```python
# operators - &, |, &&
```

```python
filt_2 = (df_tst['lname'] == 'Kaur') & (df_tst['fname'] == 'Nimrat')
df_tst.loc[filt_2, 'email']
```

```python
df_tst.loc[(df_tst['fname'] == 'Suminder') | (df_tst['fname'] == 'Nimrat'), 'email']
```

```python
df_tst.loc[~((df_tst['fname'] == 'Suminder') | (df_tst['fname'] == 'Nimrat')), 'email']
# tilde ~ can be used to reverse
```
