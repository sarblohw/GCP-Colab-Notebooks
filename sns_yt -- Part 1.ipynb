### Distribution Plot
+ Univariate Plots
+ Histogram or frequency distribution and KDE/Kernel Density Estimate plot
+ Histogram dicretesizes a continuous variable by using bins (also called intervals, classes, buckets)
+ KDE plots depict density distribution of Kernels using a line, does not discretesize the continuous variable
+ A combination of the 2 plots helps plot a distribution plot using sns

```python
sns.distplot(car_crashes.not_distracted)
```

```python
sns.distplot(car_crashes.not_distracted, kde=False)
```

```python
sns.distplot(car_crashes.not_distracted, bins=25)
#decreasing bins can reduce information while increasing too much can cause noise in the data
```

```python
sns.distplot(car_crashes.not_distracted, hist=False)
```

## Joint Plot
+ Plots distribution between 2 variables - Bivariate
+ Plots scatter plot by default and shows 2 distribution plots for each variable

```python
sns.jointplot(x='speeding', y='alcohol', data=car_crashes, kind='kde', fill=True)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=car_crashes)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=car_crashes, kind='reg')
```

```python
sns.heatmap(car_crashes.corr())
```

```python
sns.jointplot(x='speeding', y='alcohol', data=car_crashes, kind='hex')
```

```python
sns.kdeplot(car_crashes.speeding, fill=True)
```

```python
sns.kdeplot(car_crashes.alcohol)
```
